```python
country_analysis = df.groupby('Country').agg(
    Total_Attacks=('AttackType','count'),
    Total_Killed=('Killed','sum'),
    Total_Wounded=('Wounded','sum')
)
country_analysis
```

```python
# Add average casualty per attack
country_analysis['Avg_Casualty_Per_Attack'] = \
(country_analysis['Total_Killed'] + country_analysis['Total_Wounded']) / \
country_analysis['Total_Attacks']

# Sort and get top 10 countries
top10_countries = country_analysis.sort_values(
    by='Total_Attacks',
    ascending=False
).head(10)

top10_countries
```

```python
country_analysis['Total_Casualties']=country_analysis['Total_Killed'] + country_analysis['Total_Wounded']
```

```python
country_analysis
```

Normalize Total Attacks

```python
country_analysis['Norm_Attacks'] = (
    country_analysis['Total_Attacks'] - country_analysis['Total_Attacks'].min()
) / (
    country_analysis['Total_Attacks'].max() - country_analysis['Total_Attacks'].min()
)
```

Normalize Total Casualties

```python
country_analysis['Norm_Total_Casualties'] = (
    country_analysis['Total_Casualties'] - country_analysis['Total_Casualties'].min()
) / (
    country_analysis['Total_Casualties'].max() - country_analysis['Total_Casualties'].min()
)
```

Normalize Avg Casualty Per Attack

```python
country_analysis['Norm_Avg_Casualty'] = (
    country_analysis['Avg_Casualty_Per_Attack'] - country_analysis['Avg_Casualty_Per_Attack'].min()
) / (
    country_analysis['Avg_Casualty_Per_Attack'].max() - country_analysis['Avg_Casualty_Per_Attack'].min()
)
```

```python
country_analysis['Risk_Score'] = (
    0.3 * country_analysis['Norm_Attacks'] +
    0.5 * country_analysis['Norm_Total_Casualties'] +
    0.2 * country_analysis['Norm_Avg_Casualty']
)
```

```python
top10_risk = country_analysis.sort_values(
    by='Risk_Score',
    ascending=False
).head(10)

top10_risk
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.barh(
    top10_risk.index,
    top10_risk['Risk_Score']
)

plt.title("Top 10 High Risk Countries (Terrorism Risk Index)")
plt.xlabel("Risk Score")
plt.ylabel("Country")

plt.gca().invert_yaxis()
plt.show()
```

```python
df_groups=df[df['Group']!='Unknown'];
df_groups.head(2)
```

```python
group_analysis = df_groups.groupby('Group').agg(
    Total_Attacks=('AttackType','count'),
    Total_Casualties=('Total_Casualty','sum')
)

group_analysis['Avg_Casualty_Per_Attack'] = (
    group_analysis['Total_Casualties'] /
    group_analysis['Total_Attacks']
)

group_analysis = group_analysis.sort_values(
    by='Total_Attacks',
    ascending=False
)

group_analysis.head(10)
```

```python
# group_analysis.sort_values(by=['Total_Casualties','Avg_Casualty_Per_Attack'],ascending=False).head(5)

top10_deadliest_groups = group_analysis.sort_values(
    by='Total_Casualties',
    ascending=False
).head(10)

top10_deadliest_groups
```

```python
top10_intense_groups = group_analysis.sort_values(
    by='Avg_Casualty_Per_Attack',
    ascending=False
).head(10)

top10_intense_groups
```

```python
top10_active_groups = group_analysis.sort_values(
    by='Total_Attacks',
    ascending=False
).head(10)

top10_active_groups
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.barh(
    top10_deadliest_groups.index,
    top10_deadliest_groups['Total_Casualties']
)

plt.title("Top 10 Deadliest Terror Groups (By Total Casualties)")
plt.xlabel("Total Casualties")
plt.ylabel("Terror Group")

plt.gca().invert_yaxis()
plt.show()
```

TARGET TYPE ANALYSIS

```python
target_analysis = df.groupby('Target_type').agg(
    Total_Attacks=('AttackType','count'),
    Total_Casualties=('Total_Casualty','sum')
)

target_analysis['Avg_Casualty_Per_Attack'] = (
    target_analysis['Total_Casualties'] /
    target_analysis['Total_Attacks']
)

top_targets = target_analysis.sort_values(
    by='Total_Casualties',
    ascending=False
)

plt.figure(figsize=(12,6))
plt.barh(top_targets.head(10).index,
         top_targets.head(10)['Total_Casualties'])
plt.title("Top 10 Most Impacted Target Types")
plt.xlabel("Total Casualties")
plt.gca().invert_yaxis()
plt.show()
```

WEAPON TYPE ANALYSIS

```python
weapon_analysis = df.groupby('Weapon_type').agg(
    Total_Attacks=('AttackType','count'),
    Total_Casualties=('Total_Casualty','sum')
)

weapon_analysis['Avg_Casualty_Per_Attack'] = (
    weapon_analysis['Total_Casualties'] /
    weapon_analysis['Total_Attacks']
)

top_weapons = weapon_analysis.sort_values(
    by='Total_Casualties',
    ascending=False
)


plt.figure(figsize=(12,6))
plt.barh(top_weapons.head(10).index,
         top_weapons.head(10)['Total_Casualties'])
plt.title("Top 10 Deadliest Weapon Types")
plt.xlabel("Total Casualties")
plt.gca().invert_yaxis()
plt.show()
```
